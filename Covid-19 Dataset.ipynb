# Q. 1) Show the number of Confirmed, Deaths and Recovered cases in each Region.

```python
covid['Region'].value_counts()
```

# Q. 2) Remove all the records where the Confirmed Cases is Less Than 10.

```python
covid.head()
```

```python
covid = covid[covid['Confirmed'] >= 10]
```

```python
covid.head()
```

# Q. 3) In which Region, maximum number of Confirmed cases were recorded ?

```python
covid.head()
```

```python
covid.groupby('Region')['Confirmed'].sum().sort_values(ascending = False).head(10)

# to get top 10 results
```

```python
covid.groupby('Region')['Confirmed'].sum()
```

# Q. 4) In which Region, minimum number of Deaths cases were recorded ?

```python
covid.head()
```

```python
covid.groupby('Region')['Deaths'].sum().sort_values().head(50)
```
