# Q. 1) Convert the Datatype of 'Date' column to Date-Time format.

```python
london.head()
```

```python
london['date'] = pd.to_datetime(london['date'])
```

```python
london.head()
```

```python
london.info()
```

# Q. 2) Add a new column ''year'' in the dataframe, which contains years only.(B.2) Add a new column ''month'' as 2nd column in the dataframe, which contains month only.

```python
london.head()
```

```python
london['year'] = london['date'].dt.year
```

```python
london.head()
```

```python
london['month'] = london['date'].dt.month
```

```python
london.head()
```

```python
london.drop('month', axis = 1 , inplace = True)
```

```python
london.head()
```

# To add month as second index column

```python
london.insert(1, 'month' , london['date'].dt.month)
```

```python
london.head()
```

# Q. 3) Remove the columns 'year' and 'month' from the dataframe.

```python
london.head()
```

```python
london.drop(['month', 'year'], axis = 1, inplace = True)
```

```python
london.head()
```

# Q. 6) What is the Maximum & Minimum No. of Crimes recorded per area ?
