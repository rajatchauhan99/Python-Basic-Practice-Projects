# Q. 2) In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
netflix.head()
```

```python
netflix.info()
```

```python
# Converting release_date field into pandas datetime object
```

```python
netflix['Release_Date'] = pd.to_datetime(netflix['Release_Date'])
```

```python
netflix.info()
```

```python
# adding an year column 

netflix.insert(7, 'year', netflix['Release_Date'].dt.year)
```

```python
netflix.head()
```

```python
netflix.info()
```

```python
# here year column is in float lets change its type
```

```python
netflix['year'].isnull().sum()
```

```python
netflix['year'].fillna(0, inplace = True)
```

```python
netflix['year'].isnull().sum()
```

```python
netflix.info()
```

```python
netflix['year'] = netflix['year'].astype(int)
```

```python
netflix.info()
```

```python
netflix.head()
```

In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
netflix['year'].unique()
```

```python
year_data = netflix.groupby('year')['Show_Id'].count().reset_index()
```

```python
year_data
```

```python
fig, ax = plt.subplots(figsize = (15,7), dpi = 300)

sns.barplot(x = 'year', y = 'Show_Id', data = year_data, ax = ax)

ax.set_title('Bar Plot for number of movie releases per year')
```
